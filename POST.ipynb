```python
# Check missing values
print(df.isnull().sum())

# If missing values exist, fill or drop them
df = df.dropna()  # removes missing rows
```

```python
# Sleep Quality vs Stress Level
plt.figure(figsize=(6,5))
sns.barplot(data=df, x="sleep_quality", y="stress_level", palette="viridis", ci=None)
plt.title("Average Stress Level by Sleep Quality", fontsize=14, weight='bold')
plt.xlabel("Sleep Quality", fontsize=12)
plt.ylabel("Average Stress Level", fontsize=12)
```

```python
# Set the style for the plots
sns.set_style("whitegrid")
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['figure.figsize'] = (6, 6)

# Create a histogram for 'anxiety_level'
plt.figure()
sns.histplot(df['anxiety_level'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Anxiety Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Frequency')
plt.savefig('anxiety_level_histogram.png')

# Create a bar chart for 'stress_level' distribution
stress_counts = df['stress_level'].value_counts().sort_index()
plt.figure()
sns.barplot(x=stress_counts.index, y=stress_counts.values, palette='viridis')
plt.title('Distribution of Stress Levels')
plt.xlabel('Stress Level')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2], labels=['Low Stress', 'Medium Stress', 'High Stress'])
plt.savefig('stress_level_bar_chart.png')
```

```python
# Create a count plot for mental_health_history within each stress level
plt.figure()
sns.countplot(x='stress_level', hue='mental_health_history', data=df, palette='tab10')
plt.title('Mental Health History by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Count')
plt.legend(title='Mental Health History', labels=['No History', 'Has History'])
plt.savefig('mental_health_history_by_stress_count_plot.png')
```

```python
# Create a scatter plot of anxiety vs. depression colored by stress level
plt.figure(figsize=(6, 6))
sns.scatterplot(x='anxiety_level', y='depression', hue='stress_level', data=df, palette='viridis', s=60, alpha=0.7)
plt.title('Anxiety Level vs. Depression, Colored by Stress Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Depression Score')
plt.legend(title='Stress Level')
plt.savefig('anxiety_depression_scatter_plot.png')
```
